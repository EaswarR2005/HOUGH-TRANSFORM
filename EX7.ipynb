```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bird.jpeg') 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB));plt.title("Input Image")
plt.subplot(122);plt.imshow(gray_image,cmap='gray');plt.title("Gray Image")
plt.show()
```

```python
blur = cv2.GaussianBlur(gray_image, (5,5), 0)
plt.imshow(blur, cmap='gray'); plt.title("Blurred Image"); plt.show()
```

```python
low = 50
high = 150
edges = cv2.Canny(blur, low, high)
plt.imshow(edges, cmap='gray'); plt.title("Canny Edges"); plt.axis('off'); plt.show()
```

```python
lines = cv2.HoughLinesP(
    edges, 
    rho=1, 
    theta=np.pi/180, 
    threshold=50, 
    minLineLength=20, 
    maxLineGap=10
)
output = image.copy()

if lines is not None:
    print("Lines detected:", len(lines))
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 2)
else:
    print("No lines detected.")
```

```python
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title("Hough Transform Output")
plt.axis('off')
plt.show()
```
